```python
import numpy as np
import pandas as pd
import recsys as rs
```

```python
dataset = pd.read_csv('Data/train_final.csv','\t')
tr_info = pd.read_csv('Data/tracks_final.csv','\t')
tgt_pl = pd.read_csv('Data/target_playlists.csv','\t')
tgt_tr = pd.read_csv('Data/target_tracks.csv','\t')
attributes = ['artist_id', 'album', 'tags', 'playcount', 'duration']
n_min_attr = 5
n_el_sim = 20
measure = 'dot'
shrinkage = 0
min_tracks_in_playlist=10
test_percentage=20
n_tracks_toremove=5
seed=2517
```

```python
train, test, tgt_tracks, tgt_playlists = rs.split_train_test(dataset, min_tracks_in_playlist, test_percentage, n_tracks_toremove, seed)
```

```python
test_reduced = test[test['track_id'].isin(tr_info_reduced['track_id'].values)]
test_reduced
```

```python
recommendetions = np.array([])
for p in IX_tgt_playlists.values:
            avg_sims = URM[p,:].dot(S).toarray().ravel()
            top = rs.top5_outside_playlist(avg_sims, p, train_reduced, IX_tgt_playlists, IX_tgt_items)
            recommendetions = np.append(recommendetions, rs.sub_format(top))
            if (p % 1000 == 0):
                print('Recommended ' + str(p) + ' users over ' + str(IX_tgt_playlists.values.shape[0]))
print(recommendetions.shape[0])
print(tgt_pl_reduced['playlist_id'].shape[0])
print(IX_tgt_playlists.shape[0])
#recs = pd.DataFrame(data=np.array([tgt_pl_reduced['playlist_id'], recommendetions]).transpose() , index=range(IX_tgt_playlists.shape[0]), columns=['playlist_id', 'track_ids'])
recs = pd.DataFrame({'playlist_id' : IX_tgt_playlists.index.values, 'track_ids' : recommendetions})
recs
```
