```python
import numpy as np
import pandas as pd
import recsys as rs
```

```python
dataset = pd.read_csv('Data/train_final.csv','\t')
tr_info = pd.read_csv('Data/tracks_final.csv','\t')
tgt_pl = pd.read_csv('Data/target_playlists.csv','\t')
tgt_tr = pd.read_csv('Data/target_tracks.csv','\t')
attributes = ['artist_id', 'album', 'tags', 'playcount', 'duration']
n_min_attr = 1000
n_el_sim = 20
measure = 'dot'
shrinkage = 0
min_tracks_in_playlist=10
test_percentage=20
n_tracks_toremove=5
seed=2517
```

```python
tr_info_reduced = tr_info.sample(n=5000, random_state=2517, axis=0)
tr_info_reduced
```
