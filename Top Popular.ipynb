## Extracting Top Popular tracks

```python
targets = pd.DataFrame(traindata['track_id'].value_counts())
targets['tmp'] = targets.index
targets.index = np.arange(targets.shape[0])
targets.rename(columns = {'track_id':'rating', 'tmp':'track_id'}, inplace=True)
targets.head()
```

```python
top_ids = np.array_str(targets['track_id'].values[0:5])[1:-1]
top_ids
```

## Formatting as requested

```python
final = targetplaylists.copy()
final['track_ids'] = top_ids
final['track_ids'] = final['track_ids'].apply(lambda x: " ".join(x.split()))
final
```

## Saving to csv

```python
final.to_csv('Submissions/top_popular_submission.csv', index=False)
```
